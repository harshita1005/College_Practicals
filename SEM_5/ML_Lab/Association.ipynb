```python
transactions=[]
for i in range (0,7501):
    transactions.append([str(A.values[i,j]) for j in range (0,20)])
    
transactions[0]
```

```python
rule_list=apriori(transactions,min_length=2,min_support=0.003,min_confidence=0.003,min_lift=3)
result=list(rule_list)
print(result[0])
```

```python
print(len(result))
```

```python
def inspect(result):
    lhs         = [tuple(results[2][0][0])[0] for results in result]
    rhs         = [tuple(results[2][0][1])[0] for results in result]
    support    = [results[1] for results in result]
    confidence = [results[2][0][2] for results in result]
    lift       = [results[2][0][3] for results in result]
    return list(zip(lhs, rhs, support, confidence, lift))
output_DataFrame = pd.DataFrame(inspect(result), columns = ['Left_Hand_Side', 'Right_Hand_Side', 'Support', 'Confidence', 'Lift'])
output_DataFrame
```
